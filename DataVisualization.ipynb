Four Line of Code

1. import matplotlib.pyplot as plt
2. fig, ax = plt.subplot()
3. 
- ax.plot(x,y) # Line chart
- ax.bar(x,y)
- ax.barh(x,y)
- ax.scatter(x,y)
- ax.hist(x)
- ax.pie(x)

4. plt.show()

```python
x=[2,3,4,5,6,7,8,9]
y=[1,2,3,4,5,6,7,8]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x,y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(x,y)
plt.show()
```

```python
z=[1,1,1,1,1,1,12,2,2,2,2,2,2,4,4,4,4,4,4,6,6,6,6,8,8,8,9]
fig, ax = plt.subplots()
ax.hist(z)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(x)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,y, color = 'red')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,y, color = 'red',linestyle='--',marker='v',linewidth=3)
ax.set_title('Sales Nos for FY 2022')
ax.set_xlabel('Months')
ax.set_ylabel('Sales Unit Nos')
plt.show()
```
